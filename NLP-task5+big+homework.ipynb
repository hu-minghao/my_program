# 封装代码

```python
#1.第一步：合并文本，方便生成词库
def conbinetotalltext(text_dir,output_path):
    #给出训练文本所在文件夹，将其合并为一个文本文件，存储在output_path文件中
    #注意，text_dir为文件夹路径，output_path需要预先创建txt空白文件
    namelist=[x for x in os.listdir(text_dir)]
    namelist.sort(key=lambda x:int(x[:-4]))   #txt file name
    print('文本数量：%d'%len(namelist))#100
    outfilename=output_path
    outfile=open(outfilename,'a',encoding='utf-8')#a没有文件可以创建
    for i in range( len(namelist) ):
        datapath = os.path.join(text_dir, namelist[i])
        print(datapath)
        file=open(datapath,'r',encoding='utf-8')
        shutil.copyfileobj(file,outfile)
        file.close()
    outfile.close()
    with open(output_path,'r',encoding='utf-8') as f:
        text=f.read()
    return text

#第二步，利用第一步的文本及结巴分词，形成词库字典，字典中包含库中词出现的次数
def getword_dic(s):
    word_dic={}
    for i in s:
        if i.isalpha():
            word_dic[i]=word_dic.get(i,0)+1
        else:
            continue
    items = list(word_dic.items())    #把对象对象转化为列表形式，利于下面操作
    #sort()  函数用于对原列表进行排序，如果指定参数，则使用比较函数指定的比较函数
    #reverse 排序规则，reverse = True 降序， reverse = False 升序（默认）
    #key 是用来比较的参数
    print('词库大小为：%d'%len(items))
    items.sort(key=lambda x: x[1], reverse = True)
    for i in range(10):
        word, count= items[i]
        print("{0:<10}{1:>5}".format(word, count))
    return word_dic
```

```python
#第三步，计算TF_IDF矩阵函数
def get_TF_IDF(word_dic,text_dir):
    #根据文本数，词汇维度生成统计矩阵，行：文档数，列：词汇数
    #text_dir为文档所在文件夹
    word_vec_dic={}
    namelist=[x for x in os.listdir(text_dir)]
    namelist.sort(key=lambda x:int(x[:-4])) 
    word_matrix=np.zeros((len(namelist),len(word_dic)),dtype=int)
    #生成词汇表
    train_dic_list=list(word_dic.keys())
    for i in range(len(train_dic_list)):
        word_vec_dic[train_dic_list[i]]=i
    
    def count_word_in_document(file_path,word_dir,arr):
        with open(file_path,encoding='utf-8',mode='r') as f:
            text=f.read()
            for idx,item in enumerate(word_dir):
                arr[idx]=text.count(item)
            
    start=time.clock()

    for i in range(word_matrix.shape[0]):
        datapath = os.path.join(text_dir, namelist[i])
        count_word_in_document(datapath,train_dic_list,word_matrix[i])
    print('done')
    
    #记录耗时情况
    end = time.clock()
    print('Running time: %.2f Seconds'%(end-start))

    #计算TF矩阵
    TF=np.zeros((len(namelist),len(word_dic)))
    for i in range(word_matrix.shape[0]):
        TF[i]=word_matrix[i]/word_matrix[i].sum()

    #计算IDF矩阵
    count_idf=word_matrix.copy()
    #count_idf为词的基于共现矩阵的词向量，维度随文档数变化
    count_idf=count_idf.astype('bool').astype('int')
    idf_res=count_idf.sum(axis=0)
    idf=np.zeros((len(word_dic),len(namelist)),dtype='float64')
    for i in range(word_matrix.shape[1]):
        idf[i]=np.log10(100/(idf_res[i]+1))
    IDF=idf.T
    TF_IDF=TF*IDF
    return TF_IDF,word_vec_dic

```

```python
#第四步，生成各文档主题词
def get_maininfo(TF_IDF,text_dir,word_dic):
    #根据tfidf矩阵，文档文件夹，词表生成主题词
    namelist=[x for x in os.listdir(text_dir)]
    namelist.sort(key=lambda x:int(x[:-4])) 
    train_dic_list=list(word_dic.keys())
    s=pd.Series(train_dic_list,name='主题词')
    main_df=pd.DataFrame({'文档':namelist})
    main_words=[]
    for i in range(len(namelist)):
        word=[x for x in s[TF_IDF[i]>=np.percentile(TF_IDF[i],99.9)] if len(x)>1]
        main_word=','.join(word)
        main_words.append(main_word)
    main_df['主题词']=main_words
    print(main_df.head(10))
    return main_df
    
```
